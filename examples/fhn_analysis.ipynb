## Analysis on numerical solution of Fitzhugh-Nagumo model

It is important to know the rate of changes in the solution of Fitzhugh-Nagumo model as it provide insights to the speed of excitation or recovery of the excitable system. However, in this work, we would first ignore the time parameter. We would analyse on the convergence and error of the numerical methods. 

The ode23 method is used for the analysis.

# Define Fitzhugh-Nagumo model

```python
def fitzhugh_nagumo(x, y):
    alpha = 0.1
    gamma = 0.5
    epsilon = 0.01
    eq1 = (y[0] * (1 - y[0]) * (y[0] - alpha) - y[1] + 0.026) / epsilon
    eq2 = y[0] - gamma * y[1]

    return [eq1, eq2]
```

# Solve the model

```python
import solver

# Define initialisation value for numerical method
x_min = 0
x_max = 2
initial_value = [0.01, 0.01]
mesh_points = 100

problem = solver.AdaptiveMethod(
    fitzhugh_nagumo, x_min, x_max, initial_value, initial_mesh=0.1)

# Solve with adaptive method, BS23 algorithm
mesh, soln, errors = problem.ode23()

# PLot results
import matplotlib.pyplot as plt

v = []
w = []
for i in range(len(soln)):
    v.append(soln[i][0])
    w.append(soln[i][1])

plt.figure()
plt.scatter(v,w)
plt.show()
```
